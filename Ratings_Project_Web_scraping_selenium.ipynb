```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
#Lets first connect to the webdriver
driver=webdriver.Chrome("chromedriver.exe")
```

```python
#Opening the url
url='https://www.amazon.com/'
driver.get(url)

time.sleep(2)
# Asking the user to input the keywords he/she wants to search
user_inp = input('Enter the product you want to search : ')
```

```python
search_bar = driver.find_element_by_id("twotabsearchtextbox")    # Locating search_bar by id
search_bar.clear()                                               # clearing search_bar
search_bar.send_keys(user_inp)                                   # sending user input to search bar
search_button = driver.find_element_by_id("nav-search-submit-button")       # Locating search_button by id
search_button.click()                                                                # Clicking the button to start search
```

```python
# scrape all product urls
urls=[]
for i in range(0,2):#for loop for scrapping 2 pages
    for j in driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]'):
        urls.append(j.get_attribute('href'))
```

```python
urls
```

```python
len(urls)
```

```python
#defining variables
ratings_list=[]
reviews_list=[]
```

```python
#fetching ratings of the product
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        for j in driver.find_elements_by_xpath('//span[@class="a-size-base a-nowrap"]//span'):
            ratings_list.append(j.text)
    except NoSuchElementException:
        ratings_list.append('-')
```

```python
ratings_list
```

```python
len(ratings_list)
```

```python
#fetching reviews of the product
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        for j in driver.find_elements_by_xpath('//a[@class="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"]'):
            reviews_list.append(j.text)
    except NoSuchElementException:
        reviews_list.append('-')
```

```python
reviews_list
```

```python
len(reviews_list)
```
